Ecualizacin de histograma

```python
histo = np.zeros(256)
```

```python
filas = gris.shape[0]
columnas =  gris.shape[1]
salida = np.zeros((filas,columnas))
```

```python
for i in range(filas):
  for j in range(columnas):
    pixel = gris[i,j]
    histo[pixel] +=1
```

```python
plt.figure(2)
plt.plot(histo)
```

```python
pro = histo/(filas*columnas)
plt.figure(3)
plt.plot(pro)
```

```python
ecualiza = np.zeros(256)
acumulado = 0
for k in range(256):
  acumulado = pro[k]+acumulado
  ecualiza[k] = acumulado
  
```

```python
plt.figure(4)
plt.plot(ecualiza)
```
